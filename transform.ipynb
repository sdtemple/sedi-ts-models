```python
# basic
import numpy as np
import pandas as pd
```

```python
station1 = 1
station2 = 290

pseudo_count2 = 0.254 / np.sqrt(1.25)

for _ in range(station1, station2 + 1):
    if _ % 10 == 0:
        print(f'Station {_} transformed')
    table = pd.read_csv(f'cleaned/{_}_clean_data.csv')
    assert (table.sedyld.min() > 0)
    table['log_pcp'] = np.log(table.pcp + pseudo_count2)
    table['log_sedyld'] = np.log(table.sedyld)
    table.to_csv(f'cleaned/{_}_transformed_data.csv', 
                 index=False,
                 columns=['LSU_id','date','sedyld','pcp','log_sedyld','log_pcp',
                         ]
                 )
    columns = table.columns.tolist()
    columns = [col for col in columns if col not in ['sedyld','pcp','datenum']]
    y_index = columns.index('log_sedyld')
    columns[3], columns[y_index] = columns[y_index], columns[3]
    subtable = table[table['split']==0]
    subtable.to_csv(f'training/{_}.csv',index=False,columns=columns)
    subtable = table[table['split']==1]
    subtable.to_csv(f'testing/{_}.csv',index=False,columns=columns)
```
