```python
# basic
import numpy as np
import pandas as pd
```

```python
smallest_table = pd.read_csv('station_smallest_values.csv', header=None, names=['station', 'smallest'])
smallest_dict = dict(zip(smallest_table['station'], smallest_table['smallest']))
```

```python
station1 = 1
station2 = 290

pseudo_count = 1e-14 / 2
pseudo_count2 = 0.254 / np.sqrt(1.25)
# min_log_y = np.log10(min_y + pseudo_count2)
# max_log_y = np.log10(max_y + pseudo_count2)
# min_log_x = np.log10(min_x + pseudo_count)
# max_log_x = np.log10(max_x + pseudo_count)

for _ in range(station1, station2 + 1):
    if _ % 10 == 0:
        print(f'Station {_} transformed')
    table = pd.read_csv(f'cleaned/{_}_clean_data.csv')
    # columns = table.columns.tolist()
    table.sedyld = table.sedyld - smallest_dict[_]
    assert (table.sedyld.min() == 0)
    table['log10_pcp'] = np.log10(table.pcp + pseudo_count2)
    table['log10_sedyld'] = np.log10(table.sedyld + pseudo_count)
    # table['minmax_pcp'] = min_max_transform(table.pcp, min_y, max_y)
    # table['minmax_sedyld'] = min_max_transform(table.sedyld, min_x, max_x)
    # table['minmax_log_pcp'] = min_max_transform(table['log_pcp'], min_log_y, max_log_y)
    # table['minmax_log_sedyld'] = min_max_transform(table['log_sedyld'], min_log_x, max_log_x)
    table.to_csv(f'cleaned/{_}_transformed_data.csv', 
                 index=False,
                 columns=['LSU_id','date','sedyld','pcp','log10_sedyld','log10_pcp',
                          # 'minmax_sedyld','minmax_pcp','minmax_log_sedyld','minmax_log_pcp'
                         ]
                 )
    columns = table.columns.tolist()
    columns = [col for col in columns if col not in ['sedyld','pcp','datenum']]
    y_index = columns.index('log10_sedyld')
    columns[3], columns[y_index] = columns[y_index], columns[3]
    # columns = columns[:3] + ['log10_sedyld','log10_pcp'] + columns[5:]
    subtable = table[table['split']==0]
    subtable.to_csv(f'training/{_}.csv',index=False,columns=columns)
    subtable = table[table['split']==1]
    subtable.to_csv(f'testing/{_}.csv',index=False,columns=columns)
```
